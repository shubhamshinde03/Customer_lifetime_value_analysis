# Customer Lifetime Value Analysis

## Introduction

       In today's data-driven business landscape, understanding and predicting Customer behaviour is of paramount importance.Customer Lifetime Value(CLTV) analysis is a method that allows business to determine the total revenue or profit a customer can generate over the course of their relationship with the business.This metric is crucial as it helps business optimize their marketing, strategies, improve customer retention, and maximize profitability.
       CLTV analysis involves calculating the total amount of money a customer spent or is expected spend on your product and services throughout their lifetime as a customer.CLTV analysis involves RFM analysis means Recency,Frequency and Monetary.These three components are the foundation for predicting the CLTV of each customer.
       The higher the CLTV value, the more valuable a customer is to your business, as they would generate more revenue and are more likely to be loyal.CLTV analysis can inform many aspects of a business, including marketing decisions, audience targeting, and churn prediction. It can also help business identify their most valuable customers and engage them with compelling offering and incentived.
       In summary, CLTV analysis is a powerful tool that can help business understand the growth and revenue value of each customer over time,boost customer loyalty, reduce churn, and improve strategic decision-making. It's a critical metric for biulding strong customer relationships and ensuring lasting relationships with them.

## Term Explanation

### Term involved in the Customer Lifetime Value Analysis:-
#### Recency :-
How recently did the customer make a purchase? Recent customer are more likely to purchase again compared to those who haven't purchased in a while.
#### Frequency :-
How often does the customer make a purchase? Customers who make frequent purchases are more likely to continue to do so.
#### Monetary Value :- 
How much money does the customer spend on purchase? Customer who spend a lot are more likely to continue to do so.

```python
import datetime as dt
latest_date=dt.datetime(2011,12,10)
latest_date
```

```python
# Recency ----> lastest_date - last_transaction date of particular customer
# Frequency  --> Repeatation Customer_ID
# Monetaoy -->  Total w.r.t Customer_ID
```

```python
RFM_score=df_final.groupby("CustomerID").agg({"InvoiceDate":lambda x:(latest_date-x.max()).days,
                                             "Invoice_No":lambda x:x.count(),
                                             "Total":lambda x:x.sum()}).reset_index()
```

```python
RFM_score.head()
```

```python
# lets rename the columns as Recency , Frequency and Monetary ......
RFM_score.rename(columns={"InvoiceDate":"Recency","Invoice_No":"Frequency","Total":"Monetary"},inplace=True)
RFM_score.head()
```

```python
RFM_score["R_marks"]=RFM_score["Recency"].rank(ascending=False)  # ascending= false becaause we want high recency --> low marks
```

```python
RFM_score.head(10)
```

```python
# Here we create a column as F_marks that will give the frequency of customer, how frequent the customer buy things..
```

```python
RFM_score["F_marks"]=RFM_score["Frequency"].rank(ascending=True)
RFM_score.head()
```

```python
# Here we create another column as M_marks for Monetary marks , how much money customer will spend on purchase 
```

```python
RFM_score["M_marks"]=RFM_score["Monetary"].rank(ascending=True)
RFM_score.head()
```

```python
# We want the percentile of the class / column -- this will give the Rank of customer 
```

```python
RFM_score["R_rank"]=(RFM_score["R_marks"]/RFM_score["R_marks"].max())*100
RFM_score.head()
```

```python
RFM_score["F_rank"]=(RFM_score["F_marks"]/RFM_score["F_marks"].max())*100
RFM_score.head()
```

```python
RFM_score["M_rank"]=(RFM_score["M_marks"]/RFM_score["M_marks"].max())*100
RFM_score.head()
```

```python
# we are understanding the generic concept - which can be apply on different industries.
```

```python
RFM=RFM_score.copy()
RFM.head()
```

```python
RFM["RFM_final"]=0.20*RFM["R_rank"]+0.30*RFM["F_rank"]+0.50*RFM["M_rank"]
RFM.head()
```

```python
RFM=RFM.round(2)
RFM.head()
```

#### Using Elbow Method :-
- For each K, calculate the total within-cluster sum of squares(WCSS). WCSS is the sum of the squared distance between each member of the cluster and its centroid.
- The idea is that increasing the number of cluster(K) will naturally decreases the distance to data points, and therefore decreses WCSS.

```python
SSE=[]
for i in range(1,10):
    kmeans1=KMeans(max_iter=100,random_state=100,n_clusters=i)
    kmeans1.fit(final_data_scaled)
    SSE.append(kmeans1.inertia_)
    
plt.plot(range(1,10),SSE,marker="o")
plt.xlabel("No. of clustes")
plt.ylabel("Sum of Squared Error")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

#### Silhouette Method :-
- This method calculate the average distance between a sample and all other points in the same class, and the average disttance between a sample and all other points in the nearest cluster.
- The optimal numeber of cluster K is the one that maximizes the average Silhouette over a range of possible values for k.

```python
for i in range(2,9):
    kmeans2=KMeans(max_iter=100,random_state=100,n_clusters=i)
    kmeans2.fit(final_data_scaled)
    silhouette=silhouette_score(final_data_scaled,kmeans2.labels_)
    print("for {} score {}".format(i,silhouette))
```

```python
# for the season we will use month information

def cal_season(x):
    if x in [11,12,1,2]:
        return "Winter"
    elif x in [3,4]:
        return "Spring"
    elif x in [5,6,7]:
        return "Summer"
    else:
        return "Monson"
```

```python
# timing --- we use the hour informarion
def cal_timing(x):
    if x in range(22,4):
        return "Night"
    elif x in range(4,11):
        return "Morning"
    elif x in range(11,16):
        return "Afternoon"
    else:
        return "Evening"
```

```python
df_final["Season"]=df_final["Month"].apply(cal_season)
df_final["Timing"]=df_final["Month"].apply(cal_timing)
df_final.head(10)
```
